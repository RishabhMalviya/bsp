```python
env = gym.make('dm_control/humanoid-stand-v0')
env = gym.wrappers.FlattenObservation(env)

obs, info = env.reset(seed=423535)
done = False
timestep = 0
total_reward = 0
gamma = 0.99

while not done and timestep < 100_000:
    timestep += 1  # 1-indexed
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    total_reward = reward + gamma*total_reward

    done = terminated or truncated

    if timestep % 100 == 0:
        print(f'{timestep} time-steps completed')
        print(f'Current return: {total_reward}')
        print()
    
env.close()
```
